```python
%matplotlib inline
import qiskit
from qiskit import *
import matplotlib.pyplot as plt
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_multivector,plot_histogram
```

```python
qc=QuantumCircuit(1)
z=int(input("initial_state: |0> or |1> :"))
if z==0:
    init_state=[1 ,0]
elif z==1:
    init_state=[0, 1]
else:
    print("wrong initial state")
qc.initialize(init_state,0)
qc.h(0)
qc.save_statevector()
qc.draw("mpl")
```

```python
backend=AerSimulator(method='statevector')
circuit=transpile(qc,backend)
out=backend.run(circuit).result()
```

```python
state=out.get_statevector()
plot_bloch_multivector(state)
plt.show()
```

```python
count=out.get_counts()
plot_histogram(count)
```
