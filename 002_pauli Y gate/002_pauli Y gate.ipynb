```python
%matplotlib inline
import qiskit 
from qiskit import *
import matplotlib.pyplot as plt
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_bloch_multivector
```

```python
qc=QuantumCircuit(1)
z=int(input("initial_state: |0> or |1> :"))
if z==0:
    init_state = [1, 0]  # |0⟩
else:
    init_state = [0, 1]  # |1⟩
qc.initialize(init_state,0)
qc.y(0)

qc.save_statevector()
qc.draw()
qc.draw("mpl")
```

```python
backend=AerSimulator(method='statevector')
circuit=transpile(qc,backend)
out=backend.run(circuit).result()
```

```python

state=out.get_statevector(0)
plot_bloch_multivector(state)
plt.show()

```

```python
from qiskit.visualization import plot_histogram
count=out.get_counts()
plot_histogram(count)
```
